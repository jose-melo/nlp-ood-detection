```python
import numpy as np
import matplotlib.pyplot as plt
import argparse
from sklearn.datasets import load_iris
from data_depth_measures.IRW import IRW
import pandas as pd
from run import load_data
```

## Integrated weigthed rank - IRIS dataset

```python
n_samples = 1000 
n_dim = 2

X_train, X_grid, xx, yy = load_data()
```

```python
irw = IRW(n_dim=n_dim, n_samples=n_samples)
dirw = irw.score(X_train=X_train, X_test=X_grid)
```

```python
plt.contourf(xx, yy, dirw.reshape(xx.shape), levels=30, cmap='viridis')
plt.scatter(X_train[:, 0], X_train[:, 1],
            label='Train data', marker='x', c='r', s=50)
plt.colorbar()
plt.xlabel('Petal length')
plt.ylabel('Petal width')
plt.show()
```
