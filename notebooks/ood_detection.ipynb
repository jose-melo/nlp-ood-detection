### Import the NLP detector

```python
from nlp_ood_detection.detector import NLPOODDetector
import seaborn as sns
```

### Set the params

```python
from nlp_ood_detection.detector import NLPOODDetector

detector_params = {
    "dataset_in": "imdb",
    "dataset_out": "paws",
    "aggregation": "mean",
    "data_folder": "../data",
    "method": "aiirw",
    "model_name": "textattack/distilbert-base-uncased-imdb",
    "max_size": 1000,
    "threshold": 0.9,
}
detector = NLPOODDetector(**detector_params)
```

### Fit and detect!

```python
detector.fit()
score = detector.score()
prediction = detector.predict()
sns.histplot(score, kde=True)
```
